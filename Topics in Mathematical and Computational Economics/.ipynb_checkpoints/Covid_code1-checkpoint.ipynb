# Pregunta 2


```python
##########A##########
df_positivos_2=df_positivos.filter(items=['DEPARTAMENTO','METODODX','EDAD'])
df_positivos_2
```

```python
df_fallecidos_2=df_fallecidos.filter(items=['DEPARTAMENTO','EDAD_DECLARADA','SEXO'])
df_fallecidos_2
```

```python
##########B#########
df_positivos_2['Es_PruebaRapida']=df_positivos_2.apply(lambda x:0 if x['METODODX']=='PCR' else 1,axis=1)
df_positivos_2['Es_PruebaMol']=df_positivos_2.apply(lambda x:1 if x['METODODX']=='PCR' else 0,axis=1)
df_positivos_2
```

```python
df_fallecidos_2['Es_Hombre']=df_fallecidos_2.apply(lambda x:1 if x["SEXO"]=='MASCULINO' else 0,axis=1)
df_fallecidos_2['Es_Mujer']=df_fallecidos_2.apply(lambda x:0 if x["SEXO"]=='MASCULINO' else 1,axis=1)
df_fallecidos_2
```

```python
df_fallecidos_2['Es_AdultMay']=df_fallecidos_2.apply(lambda x:1 if x['EDAD_DECLARADA']>65 else 0,axis=1)
df_fallecidos_2
```

```python
df_positivos_2['Rango_Edad']=df_positivos_2.apply(lambda x: 'Joven' if x['EDAD']<20 else  ('Adulto' if x['EDAD']<65 else 'Adulto Mayor'),axis=1 )
df_positivos_2
```

# Pregunta 3

```python
##########A#########
```

```python
df_dep_positivos=df_positivos_2.groupby(['DEPARTAMENTO']).size().to_frame('casos_positivos')
df_dep_positivos
```

```python
df_dep_fallecidos=df_fallecidos_2.groupby(['DEPARTAMENTO']).size().to_frame('casos_fallecidos')
df_dep_fallecidos
```

```python
df_dep_tipo_prueba=df_positivos_2.groupby(['DEPARTAMENTO']).agg({'Es_PruebaRapida':['sum'],'Es_PruebaMol':['sum']})
df_dep_tipo_prueba.columns=['positivos_prueba_rap','positivos_prueba_mol']
df_dep_tipo_prueba
```

```python
df_dep_adul_may=df_fallecidos_2.groupby(['DEPARTAMENTO']).agg({'Es_AdultMay':['sum']})
df_dep_adul_may.columns=['fallecidos_adultos_may']
df_dep_adul_may
```

```python
df_consolidado=pd.merge(df_dep_positivos,df_dep_fallecidos,on='DEPARTAMENTO',how='inner')
df_consolidado=pd.merge(df_consolidado,df_dep_tipo_prueba,on='DEPARTAMENTO',how='inner')
df_consolidado=pd.merge(df_consolidado,df_dep_adul_may,on='DEPARTAMENTO',how='inner')
df_consolidado
```

```python
df_consolidado=df_consolidado.sort_values(by=['casos_positivos'],ascending=False)
```

```python
df_consolidado.reset_index(inplace=True)
import matplotlib.pyplot as plt 
plt.rcParams.update({'font.size': 17})

df_consolidado.plot(kind ='bar',x='DEPARTAMENTO',y='casos_positivos',legend =True, figsize=(20,5))
```
