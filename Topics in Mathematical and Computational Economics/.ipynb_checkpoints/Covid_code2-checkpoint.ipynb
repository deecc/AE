```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

plt.rcParams.update({'font.size': 17})
```

```python
# 2)
# Observamos las primeras 5 filas de datos de la base y las variables que contiene
# Lo hacemos hasta el 25 para que se vea que existen datos distintos a 'Años' en el campo 'TIEMPO EDAD' y 
# distintos a 'Perú' en el campo 'PAIS DOMICILIO'. Esto será tratado luego.
df_fallecidos.head()
```

```python
# 3)
# Procedemos a filtrar las variables señaladas de la base de datos original para mejor trabajo
df_fallecidos2 = df_fallecidos.filter(items = ["TIPO SEGURO","EDAD","SEXO","TIEMPO EDAD","ESTADO CIVIL","NIVEL DE INSTRUCCIÓN","FECHA","PAIS DOMICILIO","DEPARTAMENTO DOMICILIO","DEBIDO A (CAUSA A)"])
df_fallecidos2.head(25)
```

```python
df_fallecidos2.describe()
```

```python
# 4) 
# Primero observamos las categorías que existen en las 2 columnas a filtrar
# De "Tiempo Edad" queremos solo 'Años'
df_fallecidos2['TIEMPO EDAD'].unique()
```

```python
# De esta variable solo queremos 'Perú'
df_fallecidos2['PAIS DOMICILIO'].unique()
```

```python
# Acá filtramos las variables 'Tiempo Edad' y 'País Domicilio' 
df_fallecidos2 = df_fallecidos2[df_fallecidos2['TIEMPO EDAD'] == 'AÑOS']
```

```python
df_fallecidos2 = df_fallecidos2[df_fallecidos2['PAIS DOMICILIO'] == 'PERU']
df_fallecidos2
```

```python
# 5)
# Examinamos si la base de datos filtrada tiene missings
df_fallecidos2.isnull().sum(axis = 0)
# En esta nueva base de datos no existen missings dentro de las variables
```

```python
# 6)
# Creamos los siguientes campos pertinentes para el análisis
# a)
df_fallecidos2["FECHA_FALLECIMIENTO"] = pd.to_datetime(df_fallecidos2["FECHA"],format='%Y/%m/%d') 
df_fallecidos2["FECHA_FALLECIMIENTO"]
```

```python
# b)
df_fallecidos2['CODMES_FALLECIMIENTO'] = df_fallecidos2['FECHA_FALLECIMIENTO'].dt.strftime('%Y%m')
df_fallecidos2['CODMES_FALLECIMIENTO']
```

```python
# c)
df_fallecidos2['AÑO_FALLECIMIENTO'] = df_fallecidos2["FECHA_FALLECIMIENTO"].dt.strftime('%Y')
df_fallecidos2['AÑO_FALLECIMIENTO']
```

```python
# d)
df_fallecidos2['MES_FALLECIMIENTO'] = df_fallecidos2["FECHA_FALLECIMIENTO"].dt.strftime('%m')
df_fallecidos2['MES_FALLECIMIENTO']
```

```python
# Ahora vemos la base completa luego de lo útltimo
df_fallecidos2
```

```python
# 7)
# Creamos un nuevo campo en la base de datos y la observamos
df_fallecidos2['CAUSA_MUERTE']=df_fallecidos2.apply(lambda x:"POSIBLE CORONAVIRUS" if((any(ele in (str(x["DEBIDO A (CAUSA A)"])) for ele in ["NEUMONIA","ASFIXIA","INSUFICIENCIA RESPIRATORIA","CORONAVIRUS","RESPIRATORIA","RESPÍRATORIA","CORONAVIRUS"]) and (int(x["CODMES_FALLECIMIENTO"])>=202003))) else "OTRAS CAUSAS", axis = 1)
df_fallecidos2
```

```python
# 8)
# Revisamos la calidad de los datos para su limpieza
df_fallecidos2["EDAD"] = df_fallecidos2.apply(lambda x: int(x["EDAD"]) if (str(x["EDAD"]).isnumeric()) else np.nan, axis=1)
df_fallecidos2["DEPARTAMENTO DOMICILIO"] = df_fallecidos2.apply(lambda x: None if (str(x["DEPARTAMENTO DOMICILIO"]) ==
"SIN REGISTRO") else str(x["DEPARTAMENTO DOMICILIO"]) , axis=1)
```

```python
df_fallecidos2["EDAD"]
```

```python
# Revisamos si existen NA luego de lo anterior
df_fallecidos2.isnull().sum(axis = 0)
```

```python
# Importamos stats para hallar la moda
from scipy import stats
```

```python
# Aplicamos y hallamos la moda de la variable 'EDAD'
hogar_moda=stats.mode(df_fallecidos2["EDAD"])
hogar_moda
```

```python
# Imputamos la moda en los NA del campo requerido
df_fallecidos2["EDAD"]=df_fallecidos2["EDAD"].fillna(84)
```

```python
# Para encontrar la moda de "DEPARTAMENTO DOMICILIO", que es LIMA
df_fallecidos2["DEPARTAMENTO DOMICILIO"].describe()
```

```python
# Reemplanzando los NA por la moda encontrada en la línea anterior (LIMA)
df_fallecidos2["DEPARTAMENTO DOMICILIO"]=df_fallecidos2["DEPARTAMENTO DOMICILIO"].fillna('LIMA')
```

```python
# Comprobando si se los cambios se ejecutaron en "DEPARTAMENTO DOMICILIO"
df_fallecidos2["DEPARTAMENTO DOMICILIO"].unique()
```

```python
# Comprobando si se cambiarlos los datos en general
df_fallecidos2.isnull().sum(axis = 0)
```

```python
# 9)
campo_edad = "EDAD"
def funcEdad(row): 
    if ((row[campo_edad] >= 0) & (row[campo_edad] < 10)):   
        return 'Niño'  
    elif ((row[campo_edad] >= 10) & (row[campo_edad] < 18)):      
        return 'Adolescente'    
    elif ((row[campo_edad] >= 18) & (row[campo_edad] < 40)):     
        return 'Adulto'  
    elif ((row[campo_edad] >= 40) & (row[campo_edad] < 65)):      
        return 'Adulto Intermedio'  
    elif (row[campo_edad] >= 65):       
        return 'Adulto Mayor'
df_fallecidos2['Rango_Edad'] = df_fallecidos2.apply(funcEdad, axis=1)
```

```python
df_fallecidos2
```

### Parte 2: Parte Gráfica Análisis Exploratorio

```python
# 1)
# Creamos nuevos dataframes a partir de la base df_fallecidos2
df_fall_2019 = df_fallecidos2[(df_fallecidos2['CODMES_FALLECIMIENTO'].astype(int) >=201903)&(df_fallecidos2['CODMES_FALLECIMIENTO'].astype(int)<=201906)] 
df_fall_2019
```

```python
df_fall_2020= df_fallecidos2[(df_fallecidos2['CODMES_FALLECIMIENTO'].astype(int)>=202003)&(df_fallecidos2['CODMES_FALLECIMIENTO'].astype(int)<=202006)]
df_fall_2020
```

```python
# Graficamos según lo requerido
plt.rcParams["figure.figsize"] = (20,15)
fig, axs = plt.subplots(2, 2)
plt.subplots_adjust(left=0.125, bottom=0.1, right=0.9, top=0.9, wspace=0.2, hspace=0.2) # Por convención

# Bloque 0,0
axs[0, 0].set_title("Fallecidos 2019 Mar-Jul %")
axs[0, 0].hist(x = df_fall_2019["EDAD"],bins=30,density=True, color = 'green')
# Bloque 0,1
axs[0, 1].set_title("Fallecidos 2019 Mar-Jul Freq")
axs[0, 1].hist(x = df_fall_2019["EDAD"],bins=30, color = 'green')
# Bloque 1,0
axs[1, 0].set_title("Fallecidos 2020 Mar-Jul %")
axs[1, 0].hist(x = df_fall_2020["EDAD"],bins=30,density=True,cumulative=True, color = 'red')
# Bloque 1,1
axs[1, 1].set_title("Fallecidos 2020 Mar-Jul Freq")
axs[1, 1].hist(x = df_fall_2020["EDAD"],bins=30,cumulative=True, color = 'red')
plt.show()
```

```python
# 2) Creamos nuevos campos
df_fallecidos2["flg_2017"] = df_fallecidos2.apply(lambda x: 1 if int(x["AÑO_FALLECIMIENTO"]) == 2017 else 0, axis = 1) 
df_fallecidos2["flg_2018"] = df_fallecidos2.apply(lambda x: 1 if int(x["AÑO_FALLECIMIENTO"]) == 2018 else 0, axis = 1)
df_fallecidos2["flg_2019"] = df_fallecidos2.apply(lambda x: 1 if int(x["AÑO_FALLECIMIENTO"]) == 2019 else 0, axis = 1) 
df_fallecidos2["flg_2020"] = df_fallecidos2.apply(lambda x: 1 if int(x["AÑO_FALLECIMIENTO"]) == 2020 else 0, axis = 1) 
df_fallecidos2["flg_corona"] = df_fallecidos2.apply(lambda x: 1 if x["CAUSA_MUERTE"] == "POSIBLE CORONAVIRUS" else 0, axis = 1) 
```

```python
df_fallecidos2["flg_corona"].sum
```

```python
fallecidos_17 = df_fallecidos2.groupby(["MES_FALLECIMIENTO","DEPARTAMENTO DOMICILIO","Rango_Edad","NIVEL DE INSTRUCCIÓN"]).agg( {'flg_2017':['sum']} )
fallecidos_17.columns = ["fallecidos_17"]
fallecidos_17
```

```python
fallecidos_18 = df_fallecidos2.groupby(["MES_FALLECIMIENTO","DEPARTAMENTO DOMICILIO","Rango_Edad","NIVEL DE INSTRUCCIÓN"]).agg( {'flg_2018':['sum']} )
fallecidos_18.columns = ["fallecidos_18"]
fallecidos_18
```

```python
fallecidos_19 = df_fallecidos2.groupby(["MES_FALLECIMIENTO","DEPARTAMENTO DOMICILIO","Rango_Edad","NIVEL DE INSTRUCCIÓN"]).agg( {'flg_2019':['sum']} )
fallecidos_19.columns = ["fallecidos_19"]
fallecidos_19
```

```python
fallecidos_20 = df_fallecidos2.groupby(["MES_FALLECIMIENTO","DEPARTAMENTO DOMICILIO","Rango_Edad","NIVEL DE INSTRUCCIÓN"]).agg( {'flg_2020':['sum']} )
fallecidos_20.columns = ["fallecidos_20"]
fallecidos_20
```

```python
fallecidos_posible_coronavirus = df_fallecidos2.groupby(["MES_FALLECIMIENTO","DEPARTAMENTO DOMICILIO","Rango_Edad","NIVEL DE INSTRUCCIÓN"]).agg( {'flg_corona':['sum']} )
fallecidos_posible_coronavirus.columns = ["fallecidos_posible_coronavirus"]
fallecidos_posible_coronavirus
```

```python
# Unimos estos nuevos campos en un nuevo dataframe llamado df_resumen
df_resumen=pd.merge(fallecidos_17,fallecidos_18,on=["MES_FALLECIMIENTO","DEPARTAMENTO DOMICILIO","Rango_Edad","NIVEL DE INSTRUCCIÓN"],how = 'left')
df_resumen=pd.merge(df_resumen,fallecidos_19,on=["MES_FALLECIMIENTO","DEPARTAMENTO DOMICILIO","Rango_Edad","NIVEL DE INSTRUCCIÓN"],how = 'left')
df_resumen=pd.merge(df_resumen,fallecidos_20,on=["MES_FALLECIMIENTO","DEPARTAMENTO DOMICILIO","Rango_Edad","NIVEL DE INSTRUCCIÓN"],how = 'left')
df_resumen=pd.merge(df_resumen,fallecidos_posible_coronavirus,on=["MES_FALLECIMIENTO","DEPARTAMENTO DOMICILIO","Rango_Edad","NIVEL DE INSTRUCCIÓN"],how = 'left')
df_resumen
```

```python
# 3) Agrupamos el df_resumen por el campo 'DEPARTAMENTO DOMICILIO'
df_fall_dep = df_resumen[["fallecidos_17","fallecidos_18","fallecidos_19","fallecidos_20","fallecidos_posible_coronavirus"]].groupby("DEPARTAMENTO DOMICILIO").sum().reset_index()
df_fall_dep
```

```python
plt.rcParams["figure.figsize"] = (20,15)

df_fall_dep.plot(x="DEPARTAMENTO DOMICILIO", kind="bar", stacked=True)

x = df_fall_dep["DEPARTAMENTO DOMICILIO"]
y = df_fall_dep["fallecidos_17"] + df_fall_dep["fallecidos_18"]+df_fall_dep["fallecidos_19"]+df_fall_dep["fallecidos_20"]+df_fall_dep["fallecidos_posible_coronavirus"]

fx = 0.97
fy = 1.05
for i, v in enumerate(x):
    plt.text(i*fx, y[i]*fy, str(y[i]), color='black', fontsize = 12 , rotation= 0)
    
plt.show()
```

De este gráfico se aprecia que lo que corresponde al 2020 es más grande en la mayoría de departamentos y que el número de las defunciones por síntomas del coronavirus es relevante en todos los departamentos.

```python
# 4) Ahora agrupamos por 'MES_FALLECIMIENTO' y lo guardamos como df_fall_mes
df_fall_mes = df_resumen[["fallecidos_17","fallecidos_18","fallecidos_19","fallecidos_20","fallecidos_posible_coronavirus"]].groupby("MES_FALLECIMIENTO").sum().reset_index()
df_fall_mes
```

```python
# Graficamos
plt.rcParams["figure.figsize"] = (20,15)

x = df_fall_mes['MES_FALLECIMIENTO']
y = df_fall_mes['fallecidos_17']
y1 = df_fall_mes['fallecidos_18']
y2 = df_fall_mes['fallecidos_19']
y3 = df_fall_mes['fallecidos_20']


fig, ax = plt.subplots()
ax.set_title(" Reporte de Fallecidos por Año")
plt.plot( x,  y, marker='o', markerfacecolor='blue', markersize=12, color='skyblue', linewidth=2,linestyle='-',label="2017")
plt.plot( x,  y1, marker='x', markerfacecolor='red', markersize=12, color='red', linewidth=2,linestyle='-',label="2018")
plt.plot( x,  y2, marker='p', markerfacecolor='green', markersize=12, color='green', linewidth=2,linestyle='-',label="2019")
plt.plot( x,  y3, marker='p', markerfacecolor='purple', markersize=12, color='purple', linewidth=2,linestyle='-',label="2020")

fx = 0.99
fy = 1.03

for i, v in enumerate(x):
    plt.text(i*fx,y[i]*fy,str(y[i]), color='black', fontsize = 14 ,rotation= 0)
    plt.text(i*fx,y1[i]*fy,str(y1[i]), color='black', fontsize = 14 ,rotation= 0)
    plt.text(i*fx,y2[i]*fy,str(y2[i]), color='black', fontsize = 14 ,rotation= 0)
    plt.text(i*fx,y3[i]*fy,str(y3[i]), color='black', fontsize = 14 ,rotation= 0)
    
plt.legend(title='Año')
plt.show()
```

En este gráfico se puede observar que el 2020 tiene un gran salto respecto a los fallecidos de años previos. Este salto se le puede atribuir a la pandemia por el covid provocando más decesos de forma directa o indirecta.

```python
# 6)
df_edad=df_resumen.reset_index()
df_edad=df_edad[df_edad['MES_FALLECIMIENTO'].astype(int) >= 3]
df_edad=df_edad[df_edad['MES_FALLECIMIENTO'].astype(int) <= 6]
df_edad=df_edad.groupby(['Rango_Edad','MES_FALLECIMIENTO']).agg({"fallecidos_18":["sum"],"fallecidos_19":["sum"],"fallecidos_20":["sum"]})
df_edad.columns = df_edad.columns.droplevel(1)
df_edad=df_edad.reset_index()

df_edad.loc[df_edad['MES_FALLECIMIENTO']== "03", 'Mes']= 'Marzo'
df_edad.loc[df_edad['MES_FALLECIMIENTO']== "04", 'Mes']= 'Abril'
df_edad.loc[df_edad['MES_FALLECIMIENTO']== "05", 'Mes']= 'Mayo'
df_edad.loc[df_edad['MES_FALLECIMIENTO']== "06", 'Mes']= 'Junio'

df_edad_Niño=df_edad[df_edad['Rango_Edad']=="Niño"]
df_edad_Adolescente=df_edad[df_edad['Rango_Edad']=="Adolescente"]
df_edad_Adulto=df_edad[df_edad['Rango_Edad']=="Adulto"]
df_edad_Adulto_Intermedio=df_edad[df_edad['Rango_Edad']=="Adulto Intermedio"]
df_edad_Adulto_Mayor=df_edad[df_edad['Rango_Edad']=="Adulto Mayor"]
```

```python
df_edad_Niño.plot(kind="bar",x="Mes",y=["fallecidos_18","fallecidos_19","fallecidos_20"])
plt.title('Edad: Niño')
plt.xlabel('Mes')
plt.ylabel('Fallecidos')
df_edad_Adolescente.plot(kind="bar",x="Mes",y=["fallecidos_18","fallecidos_19","fallecidos_20"])
plt.title('Edad: Adolescente')
plt.xlabel('Mes')
plt.ylabel('Fallecidos')
df_edad_Adulto.plot(kind="bar",x="Mes",y=["fallecidos_18","fallecidos_19","fallecidos_20"])
plt.title('Edad: Adulto')
plt.xlabel('Mes')
plt.ylabel('Fallecidos')
df_edad_Adulto_Intermedio.plot(kind="bar",x="Mes",y=["fallecidos_18","fallecidos_19","fallecidos_20"])
plt.title('Edad: Adulto Intermedio')
plt.xlabel('Mes')
plt.ylabel('Fallecidos')
df_edad_Adulto_Mayor.plot(kind="bar",x="Mes",y=["fallecidos_18","fallecidos_19","fallecidos_20"])
plt.title('Edad: Adulto Mayor')
plt.xlabel('Mes')
plt.ylabel('Fallecidos')

plt.show()
```

Al comparar la cantidad de fallecidos del año 2020 en comparación de los anteriores se observa lo siguiente:
1. En los niños, el número de fallecidos de cada mes es inferior a los años anteriores
2. En los adolescentes, solo en el mes abril la cantidad de fallecidos fue superior que en el 2019.
3. En los adultos, la cantidad de fallecidos del 2020 se ha ido incrementando. En mayo y junio el número es mayor que los meses del año anterior
4. En los adultos intermedios y mayores, el número de muertos de cada mes del 2020 es superior a los años anteriores
El COVID19 tiene mayor porcentaje de fallecimiento en adultos mayores e intermedios. Por ello la diferencia entre la cantidad de decesos de este año respecto a los anteriores podría ser atribuible a la pandemia actual, sea directa o indirectamente.

```python
# 7) 
# Creamos los campos necesarios para el gráfico de torta o pie.
df_departamento = df_departamento.sort_values(by=['fallecidos_posible_coronavirus'], ascending=False)
df_departamento["fall_posible_covid_acum"] = df_departamento["fallecidos_posible_coronavirus"].cumsum()
df_departamento["fall_posible_covid_acum_freq"] = df_departamento["fall_posible_covid_acum"]/ df_departamento["fallecidos_posible_coronavirus"].sum()
df_departamento.head(10)
```

```python
df_departamento["DEPARTAMENTO DOMICILIO 2"] = df_departamento.apply(lambda x: x["DEPARTAMENTO DOMICILIO"] if (x["fall_posible_covid_acum_freq"] <= 0.90) else "OTROS",axis = 1)
df_departamento
```

```python
df_departamento2 = df_departamento[["DEPARTAMENTO DOMICILIO 2","fallecidos_posible_coronavirus"]].groupby("DEPARTAMENTO DOMICILIO 2").sum().reset_index()
df_departamento2
```

```python
x = df_departamento2["DEPARTAMENTO DOMICILIO 2"] 
y = df_departamento2["fallecidos_posible_coronavirus"]
```

```python
fig1, ax1 = plt.subplots()
ax1.pie(y,  labels=x, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')
plt.show()
```

En este gráfico se aprecia que Lima tiene casi la mitad del total de fallecidos por posible coronavirus entre marzo y junio.
